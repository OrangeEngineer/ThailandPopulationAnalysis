```python
import pandas as pd
import numpy as np
```

```python
DEC = {}
```

```python
for i in range(56,65):
    DEC[str(i)] = pd.read_csv("data/1_"+str(i)+"12.csv")
```

```python
DEC["56"]
```

```python
Total_Pop = pd.DataFrame()
Total_Pop["province_id"] = DEC["56"]["รหัสจังหวัด"]
Total_Pop["province_name"] = DEC["56"]["ชื่อจังหวัด"]
```

```python
for i in range(56,65):
    DEC[str(i)][str(int("25"+str(i))-543)] = DEC[str(i)]["ผลรวมประชากรทั้งหมด"]
```

```python
DEC["56"].iloc[:,-1].to_frame()
```

```python
for i in range(56,65):
    Total_Pop = pd.concat([Total_Pop,DEC[str(i)].iloc[:,-1].to_frame()],axis = 1)
```

```python
Total_Pop.dropna(inplace = True)
```

```python
Total_Pop['province_name'] = Total_Pop['province_name'].str.replace("จังหวัด","")
```

```python
Total_Pop
```

```python
Total_Pop.to_csv("TotalThailandPopulation2013_2021.csv",index=False)
```
