```python
import pandas as pd
import numpy as np
```

```python
DEC = {}
```

```python
for i in range(56,66):
    DEC[str(i)] = pd.read_csv("data/1_"+str(i)+"12.csv")
    DEC[str(i)].dropna(subset=['ผลรวมประชากรทั้งหมด'],inplace=True)
```

```python
DEC["65"]
```

```python
for i in range(56,66):
    DEC[str(i)]['year'] = str(int("25"+str(i))-543)
```

```python
year_2013_2022 = pd.DataFrame()
```

```python
for i in range(56,66):
    year_2013_2022 = pd.concat([year_2013_2022,DEC[str(i)]],axis=0)
```

```python
year_2013_2022[year_2013_2022['year']=='2022']
```

```python
year_2013_2022[['year']].drop_duplicates()
```

```python
year_2013_2022
```

```python
year_2013_2022['ชื่อจังหวัด'] = year_2013_2022['ชื่อจังหวัด'].str.replace("จังหวัด","")
```

```python
year_2013_2022.str.replace(",","")
```

```python
for column in year_2013_2022.columns:
    if "ชาย " in column or "หญิง " in column:
        year_2013_2022[column] = year_2013_2022[column].str.replace(",","")
```

```python
year_2013_2022['< 1 yr'] = year_2013_2022['ชาย < 1 ปี'].astype(np.int64) + year_2013_2022['หญิง < 1 ปี'].astype(np.int64)
year_2013_2022['> 100 yrs'] = year_2013_2022['ชาย > 100 ปี'] + year_2013_2022['หญิง > 100 ปี']
year_2013_2022[' 1 yr'] = year_2013_2022['ชาย 1 ปี'] + year_2013_2022['หญิง 1 ปี']
for i in range (2,101):
    year_2013_2022[str(i) + ' yrs'] = year_2013_2022['ชาย '+str(i)+' ปี'] + year_2013_2022['หญิง '+str(i)+' ปี']
```

```python
year_2013_2022.to_csv("year_2013_2022.csv",index=False)
```
